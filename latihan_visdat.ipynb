## 1. Matplotlib

```python
import matplotlib.pyplot as plt

# data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 25, 35, 40, 45, 0] # Added 0 to make the list have 12 elements to match the length of 'months'

# line chart
plt.plot(months, product_A_sales)
plt.xlabel('bulan')
plt.ylabel('jumlah penjualan')
plt.title('penjualan produk A per bulan')
plt.legend(['product A'])
plt.show()

# line chart
plt.plot(months, product_B_sales)
plt.xlabel('bulan')
plt.ylabel('jumlah penjualan')
plt.title('penjualan produk B per bulan')
plt.legend(['product B'])
plt.show()
```

### b. Kustomisasi line plot


```python
import  matplotlib.pyplot as plt

# data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

# line plot
plt.plot(months, product_A_sales, marker='o', linestyle='-', color='blue', label='Product A')
plt.plot(months, product_B_sales, marker='x', linestyle='--', color='red', label='Product B')
plt.xlabel('bulan')
plt.ylabel('penjualan perbulan')
plt.title('penjualan produk per bulan')
plt.legend()
plt.grid(True)
```

### e. subplot untuk menampilkan beberapa grafik

```python
import matplotlib.pyplot as plt

# Data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

# Line Plot
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot for Product A
axs[0].plot(months, product_A_sales, label='Product A', color='blue', marker='o')
axs[0].set_title('Penjualan Produk A Per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# Plot for Product B
axs[1].plot(months, product_B_sales, label='Product B', color='red', marker='x')
axs[1].set_title('Penjualan Produk B Per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
